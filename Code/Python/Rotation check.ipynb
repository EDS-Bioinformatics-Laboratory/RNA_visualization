```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from mpl_toolkits.mplot3d import Axes3D
```

## How to find out whether two matrices are rotations of each other

Since $W$ can be any arbitrary rotation of itself, we have to find a way to check if the solutions for $W$ are rotations of each other. We do this by checking the angle between the two rows of the matrices. The equation desribing the angle between to vectors is 
$$\cos{\theta} = \frac{u\cdot v}{\lVert u\rVert_2 \cdot \lVert v \rVert_2}$$

A Rortation matrix $R$ that rotates the space by an angle of $\theta$ is given by 
$\begin{bmatrix}\cos{\theta} & -\sin{\theta}\\ \sin{\theta} & \cos{\theta} \end{bmatrix}$

To test our rotation-check, we generate a lot of these rotation matrices with different values for $\theta$. We keep rotating an arbirtrary matrix $A$ and we see how $A$ projects the same point $v$ in space, while noting the angle between the columns of $A$.

```python
A = np.eye(2)
print('Angle between columns of A: ', np.arccos(np.dot(A.T[0],A.T[1])/(np.linalg.norm(A.T[0],2)*np.linalg.norm(A.T[1],2))))

v = [1,0]
theta = -1

for i in range(50):
    theta = np.random.uniform(2*np.pi)
    R = [[np.cos(theta),-np.sin(theta)],[np.sin(theta),np.cos(theta)]]
    B = np.matmul(A,R)
    newpoint = np.matmul(R,v)
    plt.scatter(newpoint[0], newpoint[1])
plt.show()

for i in range(5):
    theta = np.random.uniform(2*np.pi)
    R = [[np.cos(theta),-np.sin(theta)],[np.sin(theta),np.cos(theta)]]
    B = np.matmul(A,R)
    print('Angle between the rows of AR: ', np.arccos(np.dot(B.T[0],B.T[1])/(np.linalg.norm(B.T[0],2)*np.linalg.norm(B.T[1],2))))
```

This equation also takes the lengths of the vectors into account, unless we multiply one or more entire columns. See what happens if we double the values of the first column of the rotated matrix.

```python
for i in range(20):
    theta = np.random.uniform(2*np.pi)
    R = [[np.cos(theta),-np.sin(theta)],[np.sin(theta),np.cos(theta)]]
    B = np.matmul(A,R)
    newpoint = np.matmul(B,v)
    plt.scatter(newpoint[0], newpoint[1], color='red')
plt.scatter(newpoint[0], newpoint[1], color='red', label='original')
print('Angle between the rows of AR: ', np.arccos(np.dot(B.T[0],B.T[1])/(np.linalg.norm(B.T[0],2)*np.linalg.norm(B.T[1],2))))

for i in range(20):
    theta = np.random.uniform(2*np.pi)
    R = [[np.cos(theta),-np.sin(theta)],[np.sin(theta),np.cos(theta)]]
    B = np.matmul(A,R)
    B[:,0] = 3*B[:,0]
    newpoint = np.matmul(B,v)
    plt.scatter(newpoint[0], newpoint[1], color='blue')
plt.scatter(newpoint[0], newpoint[1], color='blue', label='first column doubled')
plt.legend()
    
print('Angle between the rows of AR with first column doubled: ', np.arccos(np.dot(B.T[0],B.T[1])/(np.linalg.norm(B.T[0],2)*np.linalg.norm(B.T[1],2))))
plt.show()
```

For this reason, it is beneficial to check the Euclidean norms of the columns of the two matrices.

```python
B = np.matmul(A,R)
print('Euclidean norm of columns of B before doubling: ',np.sqrt(B[0,0]**2+B[1,0]**2), np.sqrt(B[0,1]**2+B[1,1]**2))
B[:,0] = 3*B[:,0]
print('Euclidean norm of columns of B after doubling: ',np.sqrt(B[0,0]**2+B[1,0]**2), np.sqrt(B[0,1]**2+B[1,1]**2))
```

As we can see, the different rotations of $A$ project $v$ all on the unit circle that $v$ lied on originally. We can also see that the angle between the rows of $A$ is approximately equal to $\frac{\pi}{2}$ radians or $90^{\circ}$. Let's try the same for a matrix of shape $3\times 2$, as holds for $W$ is this case. Note that this time, we have three rows of matrix A and correspondingly three axes to compare.

```python
A = np.array([[1,0],[0,1],[1.5,0.5]])

v = [1,0]
theta = -1

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for i in range(50):
    theta = np.random.uniform(2*np.pi)
    R = [[np.cos(theta),-np.sin(theta)],[np.sin(theta),np.cos(theta)]]
    B = np.matmul(A,R)
    newpoint = np.matmul(B,v)
    ax.scatter(newpoint[0],newpoint[1],newpoint[2])
plt.show()

for i in range(5):
    theta = np.random.uniform(2*np.pi)
    R = [[np.cos(theta),-np.sin(theta)],[np.sin(theta),np.cos(theta)]]
    B = np.matmul(A,R)
    print('Three angles between the rows of AR: ', np.arccos(np.dot(B[0],B[1])/(np.linalg.norm(B[0],2)*np.linalg.norm(B[1],2)))
         , np.arccos(np.dot(B[0],B[2])/(np.linalg.norm(B[0],2)*np.linalg.norm(B[2],2))),
          np.arccos(np.dot(B[1],B[2])/(np.linalg.norm(B[1],2)*np.linalg.norm(B[2],2))))
```

Again, we find the same results. A rotation-check script has been written in this folder under the name 'rotation-check.py'.

## Alternative: finding a rotation matrix that translates the matrices to each other

Another way to check if two matrices $A$ and $B$ are rotations of each other is by use of the definition of a rotation matrix: A matrix $R$ is a rotation matrix if and only if $\det{R}=1$ and $R^T=R^{-1}$.
If $A$ and $B$ are rotations of each other, then $AR = B$ and thus $R=BA^{-1}$.
Let's generate some pairs of rotated matrices and compute their rotation matrix.

```python
A = np.eye(2)
A_inv = np.linalg.inv(A)
print('Angle between columns of A: ', np.arccos(np.dot(A.T[0],A.T[1])/(np.linalg.norm(A.T[0],2)*np.linalg.norm(A.T[1],2))))

plt.figure(figsize=(8,8))

for i in range(5):
    theta = np.random.uniform(2*np.pi)
    R = np.array([[np.cos(theta),-np.sin(theta)],[np.sin(theta),np.cos(theta)]])
    B = np.matmul(A,R)
    R_calc=np.matmul(B,A_inv)
    print('Det(R_%s): '%str(i+1), np.linalg.det(R_calc))
    print('R_%sT:\n'%str(i+1), R_calc.T)
    print('R_%s-1:\n'%str(i+1), np.linalg.inv(R_calc))
    print()
    plt.plot([A[0,0],0,A[0,1]],[A[1,0],0,A[1,1]],linewidth=5)
    plt.plot([B[0,0],0,B[0,1]],[B[1,0],0,B[1,1]], linewidth=3)
plt.show()
```

Of course, things get more complicated when the matrices are not just simple identity matrices. The determinant stays at roughly $1$, but the inverse of the matrix is not always computed correctly for matrices with a high condition number.

```python
fig = plt.figure(figsize=(8,8))
for i in range(4):
    A = np.random.normal(size=(2,2))
    theta = np.random.uniform(2*np.pi)
    R = np.array([[np.cos(theta),-np.sin(theta)],[np.sin(theta),np.cos(theta)]])
    B = np.matmul(A,R)
    
    A_inv = np.linalg.inv(A)
    R_calc=np.matmul(B,A_inv)
    
    print('Det(R_%s): '%str(i+1), np.linalg.det(R_calc))
    print('R_%sT:\n'%str(i+1), R_calc.T)
    print('R_%s-1:\n'%str(i+1), np.linalg.inv(R_calc))
    print()
    plt.plot([A[0,0],0,A[0,1]],[A[1,0],0,A[1,1]],linewidth=5,color=['red', 'blue', 'cyan', 'green', 'purple'][i])
    plt.plot([B[0,0],0,B[0,1]],[B[1,0],0,B[1,1]], linewidth=3,color=['red', 'blue', 'cyan', 'green', 'purple'][i])
#     ax.plot([A[0,0],0,A[0,1]],[A[1,0],0,A[1,1]], [A[2,0],0,A[2,1]],linewidth=4, color=['red', 'blue', 'cyan', 'green', 'purple'][i])
#     ax.plot([B[0,0],0,B[0,1]],[B[1,0],0,B[1,1]], [B[2,0],0,B[2,1]], linewidth=2, color=['red', 'blue', 'cyan', 'green', 'purple'][i])
plt.show()
```

When the number of dimensions rises, and when the matrices are not square anymore - as will be the case for factor loadings matrices - we have no way to find $A^{-1}$. Therefore, the least squares solution is preferred.

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
for i in range(4):
    A = np.random.normal(size=(3,2))
    theta = np.random.uniform(2*np.pi)
    R = np.array([[np.cos(theta),-np.sin(theta)],[np.sin(theta),np.cos(theta)]])
    B = np.matmul(A,R)
    
#     A_inv = np.linalg.pinv(A)
    R_calc=np.linalg.lstsq(A,B)[0]
    
    print('Det(R_%s): '%str(i+1), np.linalg.det(R_calc))
    print('R_%sT:\n'%str(i+1), R_calc.T)
    print('R_%s-1:\n'%str(i+1), np.linalg.inv(R_calc))
    print()
#     plt.plot([A[0,0],0,A[0,1]],[A[1,0],0,A[1,1]],linewidth=5,color=['red', 'blue', 'cyan', 'green', 'purple'][i])
#     plt.plot([B[0,0],0,B[0,1]],[B[1,0],0,B[1,1]], linewidth=3,color=['red', 'blue', 'cyan', 'green', 'purple'][i])
    ax.plot([A[0,0],0,A[0,1]],[A[1,0],0,A[1,1]], [A[2,0],0,A[2,1]],linewidth=4, color=['red', 'blue', 'cyan', 'green', 'purple'][i])
    ax.plot([B[0,0],0,B[0,1]],[B[1,0],0,B[1,1]], [B[2,0],0,B[2,1]], linewidth=2, color=['red', 'blue', 'cyan', 'green', 'purple'][i])
plt.show()
```

Inversing a matrix is susceptible to numerical errors. Since $R^T=R^{-1}$, it is easier to check whether $R^TR=I$.

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
for i in range(4):
    A = np.random.normal(size=(3,2))
    theta = np.random.uniform(2*np.pi)
    R = np.array([[np.cos(theta),-np.sin(theta)],[np.sin(theta),np.cos(theta)]])
    B = np.matmul(A,R)
    
#     A_inv = np.linalg.pinv(A)
    R_calc=np.linalg.lstsq(A,B)[0]
    
    print('Det(R_%s): '%str(i+1), np.linalg.det(R_calc))
    print('R_%sTR_%s:\n'%(str(i+1),str(i+1)), np.matmul(R_calc.T,R_calc))
    print()
#     plt.plot([A[0,0],0,A[0,1]],[A[1,0],0,A[1,1]],linewidth=5,color=['red', 'blue', 'cyan', 'green', 'purple'][i])
#     plt.plot([B[0,0],0,B[0,1]],[B[1,0],0,B[1,1]], linewidth=3,color=['red', 'blue', 'cyan', 'green', 'purple'][i])
    ax.plot([A[0,0],0,A[0,1]],[A[1,0],0,A[1,1]], [A[2,0],0,A[2,1]],linewidth=4, color=['red', 'blue', 'cyan', 'green', 'purple'][i])
    ax.plot([B[0,0],0,B[0,1]],[B[1,0],0,B[1,1]], [B[2,0],0,B[2,1]], linewidth=2, color=['red', 'blue', 'cyan', 'green', 'purple'][i])
plt.show()
```

As shown, we have a working function to identify whether matrices are rotations of each other! This method is more reliable and exact than the angle-and-magnitude method above, so it has been implemented as the default in rotation_check.py.
