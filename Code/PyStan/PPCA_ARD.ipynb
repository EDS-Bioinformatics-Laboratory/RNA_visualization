# PPCA in PyStan with ARD prior
In this notebook, we add a hyperparameter $\alpha$ to the ppca model. This hyperparameter allows us to evaluate which components explain most of the varriance in our model.

```python
from mpl_toolkits.mplot3d import Axes3D
import pystan
import numpy as np
import seaborn as sns
sns.set()

import os
os.sys.path.append('../Python')
from ppca import PPCA
from rotation_check import rotation_check
from utils import loadStan
```

```python
ppca_ard = loadStan('ppca_ard')
```

### Test on Iris dataset
Lets apply our algorithm on some sample data as a test!

```python
from sklearn import datasets

iris = datasets.load_iris()
iris_x = iris.data
iris_y = iris.target
```

```python
fig = plt.figure(figsize=(20,5))

for j in range(4):
    ax = fig.add_subplot(1,4,j+1)
    for i in range(3):
        ax.hist(iris_x[:,j][iris_y==i], alpha=0.5, label=['Setosa','Versicolour','Virginica'][i])
    plt.legend()
    ax.set_title(['Sepal length','Sepal width','Petal length','Petal width'][j])
plt.show()
```

```python
iris_dat = {'N': 150, 'D':4, 'M':2, 'x':np.array(iris_x)}

irisfit = ppca_ard.sampling(data=iris_dat, iter=500, chains=1)
iris_fitreturn = irisfit.extract()

best_ind = np.where(iris_fitreturn['lp__']==max(iris_fitreturn['lp__']))[0][0]
latent_data = np.mean(iris_fitreturn['z'],axis=0)
```

```python
alpha = np.mean(iris_fitreturn['alpha'],axis=0)
beta =  np.mean(iris_fitreturn['beta'],axis=0)
```

```python
best_ind = np.argmax(iris_fitreturn['lp__'])
```

Although we have mostly used the mean of all samples to parametrize our parameters, the variance relevance has a skewed distribution. I find it therefore more accurate to use the median instead of the mean of all samples. It is also possible to use the most likely sample, but this is less stable, so we will not do that.

```python
plt.figure(figsize=(10,6))
for i in range(4):
    plt.hist(iris_fitreturn['alpha'][:,i], alpha=0.4, bins=75, label='Dimension '+str(i+1))
plt.xlim((0,2))
plt.legend()
plt.show()
print('Mean: ', alpha)
lps = np.exp(iris_fitreturn['lp__'])/np.sum(np.exp(iris_fitreturn['lp__']))
print('Weighted mean: ', np.average(iris_fitreturn['alpha'], axis=0, weights=lps))
print('Median: ', np.median(iris_fitreturn['alpha'], axis=0))
print('Highest likelihood: ', iris_fitreturn['alpha'][best_ind])
plt.figure(figsize=(10,6))
for i in range(2):
    plt.hist(iris_fitreturn['beta'][:,i], alpha=0.4, bins=75, label='Dimension '+str(i+1))
plt.xlim((0,2))
plt.legend()
plt.show()
print('Mean: ', beta)
print('Weighted mean: ', np.average(iris_fitreturn['beta'], axis=0, weights=lps))
print('Median: ', np.median(iris_fitreturn['beta'], axis=0))
print('Highest likelihood: ', iris_fitreturn['beta'][best_ind])
alpha = np.median(iris_fitreturn['alpha'],axis=0)
beta =  np.median(iris_fitreturn['beta'],axis=0)
```

```python
sns.set()
fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot(1,2,1)
ax.scatter(latent_data[0,:], latent_data[1,:], c=iris_y, cmap='viridis')
ax.set_title('Iris flowers in two dimensions - best estimate')
ax = fig.add_subplot(1,2,2)
ax.bar(range(np.shape(iris_x)[1]),alpha/sum(alpha))
ax.set_xticks(list(range(np.shape(iris_x)[1])))
ax.set_xticklabels(('sepal length','sepal width','petal length','petal width'))
ax.set_title('Percentage of explained variance per attribute')
plt.show()
print('alpha: ', alpha/sum(alpha))
```

As we can see, petal length was the most important variable. The plot of petal length shows the most difference between the three groups, so this matches with our intuition. We can also see that the first latent dimension is the most important and explains most of the variance fo our data. This matches with the 2-dimensional plot, where the first dimension (on the x-axis) separates the groups most.

```python
print('Variance explained per latent dimension: ',beta/sum(beta))
```
