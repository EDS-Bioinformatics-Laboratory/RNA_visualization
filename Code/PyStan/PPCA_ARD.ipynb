# PPCA in PyStan with ARD prior
In this notebook, we add a hyperparameter $\alpha$ to the ppca model. This hyperparameter allows us to evaluate which components explain most of the varriance in our model.

```python
from mpl_toolkits.mplot3d import Axes3D
import pystan
import numpy as np
import seaborn as sns
sns.set()

import os
os.sys.path.append('../Python')
from ppca import PPCA
from rotation_check import rotation_check
from utils import loadStan
from mpl_toolkits.mplot3d import Axes3D
```

```python
ppca_multi = loadStan('ppca_ard_multi', recompile=True)
ppca_ard = loadStan('ppca_ard')
ppca = loadStan('ppca')
```

### Test on Iris dataset
Lets apply our algorithm on some sample data as a test!

```python
from sklearn import datasets

iris = datasets.load_iris()
iris_x = iris.data
iris_y = iris.target
```

```python
fig = plt.figure(figsize=(20,5))

for j in range(4):
    ax = fig.add_subplot(1,4,j+1)
    for i in range(3):
        ax.hist(iris_x[:,j][iris_y==i], alpha=0.5, label=['Setosa','Versicolour','Virginica'][i])
    plt.legend()
    ax.set_title(['Sepal length','Sepal width','Petal length','Petal width'][j])
plt.show()
```

```python
iris_dat = {'N': 150, 'D':4, 'M':3, 'x':np.array(iris_x)}

irisfit = ppca_ard.sampling(data=iris_dat, iter=500, chains=1)
iris_fitreturn = irisfit.extract()

best_ind = np.where(iris_fitreturn['lp__']==max(iris_fitreturn['lp__']))[0][0]
latent_data = np.mean(iris_fitreturn['z'],axis=0)
```

```python
alpha = np.mean(iris_fitreturn['alpha'],axis=0)
```

```python
best_ind = np.argmax(iris_fitreturn['lp__'])
```

Although we have mostly used the mean of all samples to parametrize our parameters, the variance relevance has a skewed distribution. I find it therefore more accurate to use the median instead of the mean of all samples. It is also possible to use the most likely sample, but this is less stable, so we will not do that.

```python
sns.set()
plt.figure(figsize=(10,6))
for i in range(3):
    plt.hist(iris_fitreturn['alpha'][:,i], alpha=0.4, bins=75, label='Dimension '+str(i+1))
plt.xlim((0,2))
plt.legend()
plt.show()
print('Mean: ', alpha)
lps = np.exp(iris_fitreturn['lp__'])/np.sum(np.exp(iris_fitreturn['lp__']))
print('Weighted mean: ', np.average(iris_fitreturn['alpha'], axis=0, weights=lps))
print('Median: ', np.median(iris_fitreturn['alpha'], axis=0))
print('Highest likelihood: ', iris_fitreturn['alpha'][best_ind])
```

```python
sns.set()
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(2,2,1)
ax.scatter(latent_data[0,:], latent_data[1,:], c=iris_y, cmap='viridis')
ax.set_xlabel('First latent dimension')
ax.set_ylabel('Second latent dimension')
ax.set_title('Iris flowers in two dimensions - best estimate')
ax = fig.add_subplot(2,2,2)
ax.scatter(latent_data[0,:], latent_data[2,:], c=iris_y, cmap='viridis')
ax.set_xlabel('First latent dimension')
ax.set_ylabel('Third latent dimension')
ax.set_title('Iris flowers in two dimensions - best estimate')
ax = fig.add_subplot(2,2,3)
ax.scatter(latent_data[2,:], latent_data[1,:], c=iris_y, cmap='viridis')
ax.set_xlabel('Third latent dimension')
ax.set_ylabel('Second latent dimension')
ax.set_title('Iris flowers in two dimensions - best estimate')
ax = fig.add_subplot(2,2,4)
ax.bar(range(3),alpha/sum(alpha))
ax.set_xticks(list(range(np.shape(iris_x)[1])))
# ax.set_xticklabels(('sepal length','sepal width','petal length','petal width'))
ax.set_xticklabels((1,2,3))
ax.set_title('Fraction of explained variance per latent dimension')
plt.show()
print('alpha: ', alpha)
print('normalized alpha: ', alpha/sum(alpha))
```

We can see that the second latent dimension is the most important and explains most of the variance fo our data. This matches with the 2-dimensional plots, where the second dimension separates the groups most. To check if the ARD prior indeed eliminates unnecessary latent dimensions, we generate a new simulated dataset.

```python
z = np.random.normal(np.zeros(4), np.ones(4), size=(100,4))
W = np.random.normal(size=(10,4))
x = np.matmul(W,z.T)

for i in range(10):
    plt.scatter(np.ones(100)*i, x.T[:,i])
plt.title('Spread of variables in dataset')
plt.xticks(list(range(10)),list(range(1,11)))
plt.show()
```

```python
dat = {'N': 100, 'D':10, 'M':9, 'x':x.T}

fit = ppca.sampling(data=dat, iter=250, chains=1)
fitreturn = fit.extract()
W_sim = np.mean(fitreturn['W'],axis=0)
ard_fit = ppca_ard.sampling(data=dat, iter=250, chains=1)
fitreturn = ard_fit.extract()
W_sim_ard = np.mean(fitreturn['W'],axis=0)
```

```python
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(121)
ax.set_title('without ARD prior')
for i in range(10):
    for j in range(9):
        if W_sim[i,j]>0:
            col='red'
        else:
            col='blue'
        ax.scatter(j+1,i+1, marker='s', s=abs(W_sim[i,j])**2*100,c=col)
ax.set_ylabel('Observed dimensions')
ax.set_xlabel('Latent dimensions')
ax.set_xticks(list(range(1,10)))
ax.set_yticks(list(range(1,11)))
ax = fig.add_subplot(122)
ax.set_title('with ARD prior')
for i in range(10):
    for j in range(9):
        if W_sim_ard[i,j]>0:
            col='red'
        else:
            col='blue'
        ax.scatter(j+1,i+1, marker='s', s=abs(W_sim_ard[i,j])**2*100,c=col)
ax.set_ylabel('Observed dimensions')
ax.set_xlabel('Latent dimensions')
ax.set_xticks(list(range(1,10)))
ax.set_yticks(list(range(1,11)))
plt.show()
print('alpha: ',np.mean(fitreturn['alpha'],axis=0))
```

Clearly, when the ARD prior is added, 5 of the 4 columns have values close to zero, indicating that only 4 latent variables would be enough to explain  the observed variables. The values of $\alpha$ confirm this.
We can also try to specify a prior on the observed dimensions.

```python
sns.set()
z = np.random.normal(np.zeros(4), np.ones(4), size=(100,4))
W = np.random.normal(size=(10,4))
x = np.matmul(W,z.T)
extra_observed = np.random.uniform(-5,5,size=(1,100))
x = np.concatenate((x, extra_observed, extra_observed), axis=0)

for i in range(len(x)):
    plt.scatter(np.ones(100)*i, x.T[:,i])
plt.title('Spread of variables in dataset')
plt.xticks(list(range(len(x))),list(range(1,len(x)+1)))
plt.show()
```

```python
sns.set()
fig = plt.figure(figsize=(12,15))
results = []
wth = ['without', 'with']
for lat in range(2):
    for obs in range(2):
        dat = {'N': 100, 'D':12, 'M':11, 'x':x.T, 'latent_ard':lat, 'observed_ard':obs}

        ard_fit = ppca_multi.sampling(data=dat, iter=250, chains=1)
        fitreturn = ard_fit.extract()
        W_sim = np.mean(fitreturn['W'],axis=0)
        alpha = np.mean(fitreturn['alpha'],axis=0)
        beta = np.mean(fitreturn['beta'],axis=0)

        ax = fig.add_subplot(2,2, lat*2+obs+1)
        ax.set_title(wth[lat]+' prior on latent and\n'+wth[obs]+' prior on observed dimensions')
        for i in range(12):
            for j in range(11):
                if W_sim[i,j]>0:
                    col='red'
                else:
                    col='blue'
                ax.scatter(j+1,i+1, marker='s', s=abs(W_sim[i,j])**2*100,c=col)
        ax.set_ylabel('Observed dimensions')
        ax.set_xlabel('Latent dimensions')
        ax.set_xticks(list(range(1,12)))
        ax.set_yticks(list(range(1,13)))
        
        print(wth[lat],' prior on latent and ',wth[obs],' prior on observed dimensions done!')
        if lat==1:
            print('alpha: ',alpha)
        if obs==1:
            print('beta: ', beta)
    
plt.show()
        
```

Indeed, some observed dimensions are now repressed, but not the correct ones, and the latent repression is a bit messed up now, so let's not do this anymore.
