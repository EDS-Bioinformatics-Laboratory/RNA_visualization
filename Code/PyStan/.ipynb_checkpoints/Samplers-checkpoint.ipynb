### Metropolis-Hastings

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

Function proportional to the probability density function:

```python
def prob_func(x):
    if x<-1:
        return 16-x**2
    elif x>2:
        return 9-x**2
    else:
        return x**3+1
```

```python
x_points = np.linspace(-4,3,1000)
plt.plot(x_points,list(map(prob_func,x_points)))
plt.plot([-4.5,3.5],[0,0],c='red')
plt.plot([0,0],[0,15.5],c='black')
plt.show()
```

Sampling:

```python
samples = []
x_old = 1
for i in range(1000000):
    while True:
        x_new = np.random.normal(x_old,5)
        alpha = prob_func(x_new)/prob_func(x_old)
        if alpha<np.random.uniform():
            continue
        else:
            x_old = x_new
            samples.append(x_new)
            break
```

Result:

```python
plt.hist(samples, bins=125, density=True)
plt.plot(np.linspace(-4,3,50),list(map(prob_func,np.linspace(-4,3,50)))/sum(list(map(prob_func,np.linspace(-4,4,50)))))
plt.show()
```

```python
print('E[P(x)]: ',np.mean(samples))
```

### Hamiltonian Monte Carlo
This time we use a Gaussian as a probability density function of interest.

```python
from scipy.stats import norm
```

```python
def d_log_joint(theta, d=0.00000001):
    d_theta = norm.logpdf(theta+d)-norm.logpdf(theta-d)
    d_t = d*2
    return -d_theta/d_t

def leapfrog(theta, r, epsilon):
    senor_r = r + 0.5*epsilon*d_log_joint(theta)
    senor_theta = theta + epsilon*senor_r
    senor_r = senor_r + 0.5*epsilon*d_log_joint(senor_theta)
    return senor_theta, senor_r

def prob(theta, r):
    return np.exp(logprob(theta) - 0.5 * r * r)

def logprob(theta):
    return norm.logpdf(theta)
```

```python
thetas = [-3]
r = [np.random.normal()]
epsilon=0.001
L = 1000
M = 1000
thetas_extra = [0]
r_extra = r.copy()
accepted = [True]

for m in range(M):
    r0 = np.random.normal(0,1)
    senor_theta = thetas[-1]
    senor_r = r0
    
    for i in range(L):
        senor_theta, senor_r = leapfrog(senor_theta, senor_r, epsilon)
    
    thetas_extra.append(senor_theta)
    r_extra.append(senor_r)
    
    alpha = prob(senor_theta, senor_r) / prob(thetas[-1], r0)
    if np.random.uniform()<alpha:
        theta_m = senor_theta
        r_m = -senor_r
        accepted.append(True)
    else:
        theta_m = thetas[-1]
        r_m = r[-1]
        accepted.append(False)
    thetas.append(theta_m)
    r.append(r_m)
```

```python
plt.hist(thetas, density=True, bins=20, label='accepted samples')
plt.hist(thetas_extra, density=True, bins=20, alpha=0.4, label='all samples')
plt.plot(np.linspace(-3,3,100), norm.pdf(np.linspace(-3,3,100)), linestyle='--', label='actual distribution')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(5,15))
for i in range(100):
    if accepted[i]:
        plt.scatter(thetas_extra[i], -i, c='blue')
        plt.plot([thetas_extra[i], thetas_extra[i]-r_extra[i]], [-i,-i], c='blue')
    else:
        plt.scatter(thetas_extra[i], -i, c='red')
        plt.plot([thetas_extra[i], thetas_extra[i]-r_extra[i]], [-i,-i], c='red')
#         plt.scatter(thetas_extra[i-1], -i, c='green')
#         plt.plot([thetas_extra[i-1], thetas_extra[i-1]-r_extra[i-1]], [-i,-i], c='green')
plt.yticks([0,-50,-100],[0,50,100])
plt.ylabel('step')
plt.show()
```

### Naive NUTS with L-determination

```python
def buildtree(theta, r, u, v, j, epsilon):
    if j == 0:
        senor_theta, senor_r = leapfrog(theta, r, v*epsilon)
        if u <= np.exp(norm.logpdf(senor_theta) - 0.5*senor_r*senor_r):
            C_add = {(senor_theta, senor_r)}
        else:
            C_add = set()
        s_tmp = int(norm.logpdf(senor_theta) - 0.5*senor_r*senor_r > np.log(u) - delta_max)
        return senor_theta, senor_r, senor_theta, senor_r, C_add, s_tmp
    
    else:
        theta_min, r_min, theta_plus, r_plus, C_add, s_tmp = buildtree(theta, r, u, v, j-1, epsilon)
        if v == -1:
            theta_min, r_min, _, _, C_add_add, s_tmp_tmp = buildtree(theta_min, r_min, u, v, j-1, epsilon)
        else:
            _, _, theta_plus, r_plus, C_add_add, s_tmp_tmp = buildtree(theta_plus, r_plus, u, v, j-1, epsilon)
        
        s_tmp = s_tmp * s_tmp_tmp * int((theta_plus - theta_min)*r_min >= 0) * int((theta_plus - theta_min)*r_plus >= 0)
        C_add = C_add.union(C_add_add)
        return theta_min, r_min, theta_plus, r_plus, C_add, s_tmp  
```

```python
thetas = [0]
r = [np.random.normal()]
epsilon=0.01
M = 10000
delta_max = 1000

for m in range(M):
    r0 = np.random.normal(0,1)
    u = np.random.uniform(0,np.exp(norm.logpdf(thetas[-1])-0.5*r0*r0))
    theta_min = thetas[-1]
    theta_plus = thetas[-1]
    r_min = r0
    r_plus = r0
    j = 0
    s = 1
    C = {(thetas[-1], r0)}
    
    while s==1:
        v = np.random.choice([-1,1])
        if v == -1:
            theta_min, r_min, _, _, C_add, s_tmp = buildtree(theta_min, r_min, u, v, j, epsilon)
        else:
            _, _, theta_plus, r_plus, C_add, s_tmp = buildtree(theta_plus, r_plus, u, v, j, epsilon)
            
        if s_tmp == 1:
            C = C.union(C, C_add)
            
            
        s = s_tmp * int((theta_plus-theta_min)*r_min>=0) * int((theta_plus-theta_min)*r_plus>=0)
        
        j+=1
    sample = list(C)[np.random.choice(range(len(C)))]
    thetas.append(sample[0])
    r.append(sample[1])
```

```python
plt.hist(thetas, bins=25, density=True, label='samples')
plt.plot(np.linspace(-2.5,2.5,100), norm.pdf(np.linspace(-2.5,2.5,100)), linestyle='--', label='actual distribution')
plt.legend()
plt.show()
```

### More efficient NUTS-sampler with L-determination

```python
def buildtree2(theta, r, u, v, j, epsilon):
    if j == 0:
        senor_theta, senor_r = leapfrog(theta, r, v*epsilon)
        senor_n = int(u <= prob(senor_theta, senor_r))
        s_tmp = int((logprob(senor_theta) - 0.5*senor_r*senor_r) > (np.log(u) - delta_max))
        
        return senor_theta, senor_r, senor_theta, senor_r, senor_theta, senor_n, s_tmp
    
    else:
        theta_min, r_min, theta_plus, r_plus, senor_theta, senor_n, s_tmp = buildtree2(theta, r, u, v, j-1, epsilon)
        if s_tmp == 1:
            if v == -1:
                theta_min, r_min, _, _, senor_senor_theta, senor_senor_n, s_tmp_tmp = buildtree2(theta_min, r_min, u, v, j-1, epsilon)
            else:
                _, _, theta_plus, r_plus, senor_senor_theta, senor_senor_n, s_tmp_tmp = buildtree2(theta_plus, r_plus, u, v, j-1, epsilon)
            if senor_n + senor_senor_n <= 0:
                senor_theta = senor_senor_theta
            elif np.random.uniform() < senor_senor_n/(senor_n + senor_senor_n):
                senor_theta = senor_senor_theta
            s_tmp = s_tmp_tmp * int((theta_plus - theta_min)*r_min >= 0) * int((theta_plus - theta_min)*r_plus >= 0)
            senor_n = senor_n + senor_senor_n
            
        return theta_min, r_min, theta_plus, r_plus, senor_theta, senor_n, s_tmp  
```

```python
thetas = [0]
epsilon=0.1
M = 1000
delta_max = 100

for m in range(M):
    r0 = np.random.normal(0,1)
    u = np.random.uniform(0,prob(thetas[-1], r0))
    theta_min = thetas[-1]
    theta_plus = thetas[-1]
    thetas.append(thetas[-1])
    r_min = r0
    r_plus = r0
    j = 0
    n = 1
    s = 1
    
    while s==1:
        v = np.random.choice([-1,1])
        if v == -1:
            theta_min, r_min, _, _, senor_theta, senor_n, s_tmp = buildtree2(theta_min, r_min, u, v, j, epsilon)
        else:
            _, _, theta_plus, r_plus, senor_theta, senor_n, s_tmp = buildtree2(theta_plus, r_plus, u, v, j, epsilon)
            
        if s_tmp == 1:
            if np.random.uniform() < senor_n/n:
                thetas[-1] = senor_theta
#         print(senor_theta)
        n = n + senor_n
        s = s_tmp * int((theta_plus-theta_min)*r_min >= 0) * int((theta_plus-theta_min)*r_plus >= 0)
        j = j + 1
```

```python
plt.hist(thetas, bins=50, density=True, label='samples')
plt.plot(np.linspace(-2.5,2.5,100), norm.pdf(np.linspace(-2.5,2.5,100)), linestyle='--', label='actual distribution')
plt.legend()
plt.show()
```

### Hamiltonian Monte Carlo with Dual Averaging for $\epsilon$-determination

```python
def findReasonableEpsilon(theta):
    epsilon = 1
    r = np.random.normal()
    senor_theta, senor_r = leapfrog(theta, r, epsilon)
    alpha = (2 * int((prob(senor_theta, senor_r)/prob(theta, r))>0.5)) - 1
    while (prob(senor_theta, senor_r)/prob(theta, r))**alpha > 2**(-alpha):
        epsilon = (2**alpha) * epsilon
        senor_theta, senor_r = leapfrog(theta, r, epsilon)
    return epsilon  
```

```python
thetas = [0]
delta = 1.0
lmda = 0.25
M = 5000
M_adapt = 1000

epsilon = [findReasonableEpsilon(thetas[-1])]
mu = np.log(10*epsilon[-1])
epsilon_bar = 1
H = 0
gamma = 0.05
t = 10
k = 0.75

for m in range(1,M+1):
    r0 = np.random.normal(0,1)
    thetas.append(thetas[-1])
    senor_theta = thetas[-2]
    senor_r = r0
    L_m = max(1, np.round(lmda/epsilon[-1]))
#     print(L_m)
    for i in range(int(L_m)):
        senor_theta, senor_r = leapfrog(senor_theta, senor_r, epsilon[-1])
    
    alpha = prob(senor_theta, senor_r)/prob(thetas[-2], r0)
    if alpha > np.random.uniform():
        thetas[-1] = senor_theta
        r_m = -senor_r
        
    if m <= M_adapt:
        H = (1 - (1/(m+t)))*H + (1/(m+t))*(delta-alpha)
        epsilon.append(np.exp(mu - (np.sqrt(m)/gamma)*H))
        epsilon_bar = np.exp((m**-k)*np.log(epsilon[-1]) + (1-(m**-k))*np.log(epsilon_bar))
    else:
        epsilon.append(epsilon_bar)
```

```python
plt.hist(thetas, bins=50, density=True, label='samples')
plt.plot(np.linspace(-2.5,2.5,100), norm.pdf(np.linspace(-2.5,2.5,100)), linestyle='--', label='actual distribution')
plt.legend()
plt.show()
```

### No-U-Turn Sampler with Dual Averaging

```python
def buildtree3(theta, r, u, v, j, epsilon, theta0, r0):
    if j == 0:
        senor_theta, senor_r = leapfrog(theta, r, v*epsilon)
        senor_n = int(u <= prob(senor_theta, senor_r))
        s_tmp = int(np.exp(logprob(senor_theta)+delta_max - 0.5 * senor_r * senor_r) > u)
        
        return senor_theta, senor_r, senor_theta, senor_r, senor_theta, senor_n, s_tmp, min(1, np.exp(logprob(senor_theta)-0.5*senor_r*senor_r - logprob(theta0)+ 0.5*r0*r0)), 1
    
    else:
        theta_min, r_min, theta_plus, r_plus, senor_theta, senor_n, s_tmp, senor_alpha, senor_n_alpha = buildtree3(theta, r, u, v, j-1, epsilon, theta0, r0)
        if s_tmp == 1:
            if v == -1:
                theta_min, r_min, _, _, senor_senor_theta, senor_senor_n, s_tmp_tmp, senor_senor_alpha, senor_senor_n_alpha = buildtree3(theta_min, r_min, u, v, j-1, epsilon, theta0, r0)
            else:
                _, _, theta_plus, r_plus, senor_senor_theta, senor_senor_n, s_tmp_tmp, senor_senor_alpha, senor_senor_n_alpha = buildtree3(theta_plus, r_plus, u, v, j-1, epsilon, theta0, r0)
            if senor_n + senor_senor_n == 0:
                senor_theta = senor_senor_theta
            elif senor_senor_n/(senor_n + senor_senor_n) > np.random.uniform():
                senor_theta = senor_senor_theta
            
            senor_alpha = senor_alpha + senor_senor_alpha
            senor_n_alpha = senor_n_alpha + senor_senor_n_alpha
            s_tmp = s_tmp_tmp * int((theta_plus - theta_min)*r_min >= 0) * int((theta_plus - theta_min)*r_plus >= 0)
            senor_n = senor_n + senor_senor_n
            
        return theta_min, r_min, theta_plus, r_plus, senor_theta, senor_n, s_tmp, senor_alpha, senor_n_alpha
```

```python
thetas = [0]
delta = 0.5
M = 1000
M_adapt = int(M/2)

epsilon = findReasonableEpsilon(thetas[-1])
mu = np.log(10*epsilon)
epsilon_bar = 1
H = 0
gamma = 0.05
t = 10
k = 0.75

for m in range(1,M+1):
    r0 = np.random.normal(0,1)
    u = np.random.uniform(0, prob(thetas[-1], r0))
    theta_min = thetas[-1]
    theta_plus = thetas[-1]
    r_min = r0
    r_plus = r0
    j = 0
    thetas.append(thetas[-1])
    n = 1
    s = 1
    
    while s==1:
        v = np.random.choice([-1,1])
        if v == -1:
            theta_min, r_min, _, _, senor_theta, senor_n, s_tmp, alpha, n_alpha = buildtree3(theta_min, r_min, u, v, j, epsilon, thetas[-2], r0)
        else:
            _, _, theta_plus, r_plus, senor_theta, senor_n, s_tmp, alpha, n_alpha = buildtree3(theta_plus, r_plus, u, v, j, epsilon, thetas[-2], r0)
            
        if s_tmp == 1:
            if np.random.uniform() < senor_n/n:
                thetas[-1] = senor_theta
            
        n = n + senor_n
        s = s_tmp * int((theta_plus-theta_min)*r_min >= 0) * int((theta_plus-theta_min)*r_plus >= 0)
        j = j + 1
        
    if m <= M_adapt:
        H = (1 - (1/(m+t)))*H + (1/(m+t))*(delta-(alpha/n_alpha))
        epsilon = np.exp(mu - (np.sqrt(m)/gamma)*H)
        epsilon_bar = np.exp((m**-k)*np.log(epsilon) + (1-(m**-k))*np.log(epsilon_bar))
    else:
        epsilon = epsilon_bar
```

```python
thetas = [0]
epsilon=0.1
M = 1000
delta_max = 100

for m in range(M):
    r0 = np.random.normal(0,1)
    u = np.random.uniform(0,prob(thetas[-1], r0))
    theta_min = thetas[-1]
    theta_plus = thetas[-1]
    thetas.append(thetas[-1])
    r_min = r0
    r_plus = r0
    j = 0
    n = 1
    s = 1
    
    while s==1:
        v = np.random.choice([-1,1])
        if v == -1:
            theta_min, r_min, _, _, senor_theta, senor_n, s_tmp = buildtree2(theta_min, r_min, u, v, j, epsilon)
        else:
            _, _, theta_plus, r_plus, senor_theta, senor_n, s_tmp = buildtree2(theta_plus, r_plus, u, v, j, epsilon)
            
        if s_tmp == 1:
            if np.random.uniform() < senor_n/n:
                thetas[-1] = senor_theta
#         print(senor_theta)
        n = n + senor_n
        s = s_tmp * int((theta_plus-theta_min)*r_min >= 0) * int((theta_plus-theta_min)*r_plus >= 0)
        j = j + 1
```

```python
plt.hist(thetas, density=True, label='samples')
plt.plot(np.linspace(-2.5,2.5,100), norm.pdf(np.linspace(-2.5,2.5,100)), linestyle='--', label='actual distribution')
plt.legend()
plt.show()
```
