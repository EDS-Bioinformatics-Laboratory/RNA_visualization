# Hierarchical model (still under construction)
Hello, today we will try to recreate the hierarchical model as described in *Bishop & Tipping (1998)* in PyStan! In the original model, an *interactive* workflow is presented in which the user determines the number of clusters of latent data and locates their locations for the algorithm. To mimic this approach, we execute the steps of the algorithm individually, while passing the number of clusters and their approximate locations (through a $k$-means clustering algorithm) to the next step.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pystan
import numpy as np
import seaborn as sns
from scipy.stats import norm
import itertools
from sklearn.cluster import KMeans
from sklearn import datasets
import os
import pickle
from utils import loadStan
sns.set()
```

```python
ppca_model = loadStan('ppca')
moppcas_model = loadStan('moppcas')
gmm_model = loadStan('gmm')
```

## Generating data
Data was generated according to the description in Bishop & Tippings paper.

```python
N=450
M=2
D=3
K=3

sig = 3

mu = np.random.normal(0,1.6,size=(K,D))
mu[np.random.randint(K)]+=np.random.uniform(2*sig,3*sig, size=D)

sigma = np.random.uniform(sig/2,sig,size=(D))
flat_dim = np.random.randint(D)
sigma[flat_dim]=sigma[flat_dim]/5
cov = np.diag(sigma)

while True:      # make sure to not have a negligible cluster
    theta = np.random.dirichlet(np.ones(K))
    if np.all(theta>(1.0/(2*K))):
        break
        
ind = np.ones(N)
x = np.zeros((N,D))

for n in range(N):
    opt = np.random.choice(list(range(K)), p=theta)
    ind[n] = opt
    newentry = np.random.multivariate_normal(mu[opt], cov)
    x[n,:] = newentry


    
sns.set()
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(121, projection='3d')
ax.scatter(x[:,0],x[:,1],x[:,2], c=ind, cmap='viridis')
# ax.scatter(mu[:,0], mu[:,1], mu[:,2])
ax.set_title('Observed data')
ax.set_xlabel('Gene 1')
ax.set_ylabel('Gene 2')
ax.set_zlabel('Gene 3')
ax = fig.add_subplot(122, projection='3d')
ax.scatter(x[:,0],x[:,1],x[:,2], cmap='viridis')
# ax.scatter(mu[:,0], mu[:,1], mu[:,2])
ax.set_title('Observed data')
ax.set_xlabel('Gene 1')
ax.set_ylabel('Gene 2')
ax.set_zlabel('Gene 3')
plt.show()
plt.show()
```

## Zeroth level: PPCA

```python
ppca_dat = {'N':N, 'M':M, 'D':D, 'x':x}
fit = ppca_model.sampling(data=ppca_dat, iter=200, chains=1)
fitreturn = fit.extract()

best_ind = np.where(fitreturn['lp__']==max(fitreturn['lp__']))[0][0]
```

```python
W_sim = np.mean(fitreturn['W'], axis=0)
sigma_sim = np.mean(fitreturn['sigma'], axis=0)
mu_sim = np.mean(fitreturn['mu'], axis=0)
z_sim = fitreturn['z'][best_ind]
```

```python
plt.scatter(z_sim[0,:],z_sim[1,:])
plt.show()
```

## First level: Mixture model on the latent space

```python
K_inter = 2
kmeans_init = KMeans(K_inter).fit(z_sim.T)

gmm_dat = {'N': N,'K': K_inter, 'D':M, 'y':z_sim.T}
fit = gmm_model.sampling(data=gmm_dat, chains=1, iter=100, init=[{'mu':kmeans_init.cluster_centers_}])
fit_ext = fit.extract()
best_ind = np.where(fit_ext['lp__']==max(fit_ext['lp__']))[0][0]
```

```python
z_sim_gmm = np.mean(fit_ext['z'],axis=0)
theta_sim = np.mean(fit_ext['theta'],axis=0)
mu_sim = np.mean(fit_ext['mu'],axis=0)
sigmas_sim = np.mean(fit_ext['sigma'],axis=0)
z_sim_norm = (np.exp(z_sim_gmm).T/sum(np.exp(z_sim_gmm).T))
```

```python
z_cats = []
for n in range(N):
    for k in range(K_inter):
        if z_sim_norm.T[n,k] == max(z_sim_norm.T[n]):
            z_cats.append(k)
            break
z_cats = np.array(z_cats)
```

```python
fig = plt.figure(figsize=(18,6))
ax=fig.add_subplot(1,3,1)
ax.scatter(z_sim[0,:],z_sim[1,:])
title = ax.set_title("Unlabeled latent data of cluster 2")
plt.setp(title, color='black') 
ax = fig.add_subplot(1,3,2)
ax.scatter(z_sim.T[:,0],z_sim.T[:,1], c=z_cats, cmap='viridis')
# ax.scatter(kmeans_init.cluster_centers_[:,0],kmeans_init.cluster_centers_[:,1])
# ax.scatter(mu_sim[:,0],mu_sim[:,1])
title = ax.set_title("Guessed sub-clusters")
plt.setp(title, color='black')
ax = fig.add_subplot(133, projection='3d')
ax.scatter(x[:,0],x[:,1],x[:,2], c=z_cats, cmap='viridis')
# ax.scatter(mu[:,0], mu[:,1], mu[:,2])
ax.set_title('Clusters in observed space')
ax.set_xlabel('Gene 1')
ax.set_ylabel('Gene 2')
ax.set_zlabel('Gene 3')
plt.show()
plt.show()
```

## Second level: PPCA on the individual clusters

```python
returns = []
best_inds = []
for k in range(K_inter):
    ppca_dat = {'N':sum(z_cats==k), 'M':M, 'D':D, 'x':x[z_cats==k]}
    fit = ppca_model.sampling(data=ppca_dat, iter=200, chains=1)
    fitreturn2 = fit.extract()
    returns.append(fitreturn2)

    best_ind2 = np.where(fitreturn['lp__']==max(fitreturn['lp__']))[0][0]
    best_inds.append(best_ind2)
```

```python
fig = plt.figure(figsize=(K_inter*6, 6))
for k in range(K_inter):
    W_sim2 = np.mean(returns[k]['W'], axis=0)
    sigma_sim2 = np.mean(returns[k]['sigma'], axis=0)
    mu_sim2 = np.mean(returns[k]['mu'], axis=0)
    z_sim2 = returns[k]['z'][best_inds[k]]

    ax = fig.add_subplot(1,K_inter,k+1)
    ax.scatter(z_sim2[0,:],z_sim2[1,:])
    ax.set_title('Latent space of cluster '+str(k+1))
plt.show()
```

## Third level: Finding clusters in the individual latent spaces
Cluster 1 does not show an underlying structure (and it really doesn't exist out of different components) , so we'll leave it alone for now. We do see hints of underlying sub-clusters in cluster 2, so we'll try to extract those!

```python
clus = 1    # selection of cluster to further analyze
z_sim3 = returns[clus]['z'][best_inds[clus]]

K_last = 2
kmeans_init = KMeans(K_last).fit(z_sim3.T)

gmm_dat = {'N': np.shape(z_sim3)[1],'K': K_last, 'D':M, 'y':z_sim3.T}
fit = gmm_model.sampling(data=gmm_dat, chains=1, iter=100, init=[{'mu':kmeans_init.cluster_centers_}])
fit_ext3 = fit.extract()
best_ind3 = np.where(fit_ext3['lp__']==max(fit_ext3['lp__']))[0][0]
```

```python
z_sim_gmm3 = np.mean(fit_ext3['z'],axis=0)
theta_sim3 = np.mean(fit_ext3['theta'],axis=0)
mu_sim3 = np.mean(fit_ext3['mu'],axis=0)
sigmas_sim3 = np.mean(fit_ext3['sigma'],axis=0)
z_sim_norm3 = (np.exp(z_sim_gmm3).T/sum(np.exp(z_sim_gmm3).T))
```

```python
z_cats3 = []
for n in range(np.shape(z_sim3)[1]):
    for k in range(K_last):
        if z_sim_norm3.T[n,k] == max(z_sim_norm3.T[n]):
            z_cats3.append(k)
            break
z_cats3 = np.array(z_cats3)

i=0
z_cats_total = z_cats.copy()
for n in range(N):
    if z_cats_total[n] == clus:
        if clus == 1:
            z_cats_total[n]+=z_cats3[i]
        if clus == 0:
            z_cats_total[n]-=z_cats3[i]
        i+=1
z_cats_total[z_cats_total==-1]=2
```

```python
fig = plt.figure(figsize=(18,6))
ax=fig.add_subplot(1,3,1)
ax.scatter(z_sim3[0,:],z_sim3[1,:])
title = ax.set_title("Unlabeled latent data")
plt.setp(title, color='black') 
ax = fig.add_subplot(1,3,2)
ax.scatter(z_sim3.T[:,0],z_sim3.T[:,1], c=z_cats3, cmap='viridis')
# ax.scatter(kmeans_init.cluster_centers_[:,0],kmeans_init.cluster_centers_[:,1],c='red')
# ax.scatter(mu_sim[:,0],mu_sim[:,1], c='red')
title = ax.set_title("Guessed clusters")
plt.setp(title, color='black')
ax = fig.add_subplot(133, projection='3d')
ax.scatter(x[:,0],x[:,1],x[:,2], c=z_cats_total, cmap='viridis')
# ax.scatter(mu[:,0], mu[:,1], mu[:,2])
ax.set_title('Clusters in observed space')
ax.set_xlabel('Gene 1')
ax.set_ylabel('Gene 2')
ax.set_zlabel('Gene 3')
plt.show()
```

```python
K_total = 3

returns_final = []
best_inds_final = []
for k in range(K_total):
    
    ppca_dat = {'N':sum(z_cats_total==k), 'M':M, 'D':D, 'x':x[z_cats_total==k]}
    fit = ppca_model.sampling(data=ppca_dat, iter=200, chains=1)
    fitreturn_final = fit.extract()
    returns_final.append(fitreturn_final)

    best_ind_final = np.where(fitreturn_final['lp__']==max(fitreturn_final['lp__']))[0][0]
    best_inds_final.append(best_ind_final)
```

```python
fig = plt.figure(figsize=(K_total*6, 6))
for k in range(K_total):
    W_sim_final = np.mean(returns_final[k]['W'], axis=0)
    sigma_sim_final = np.mean(returns_final[k]['sigma'], axis=0)
    mu_sim_final = np.mean(returns_final[k]['mu'], axis=0)
    z_sim_final = returns_final[k]['z'][best_inds_final[k]]

    ax = fig.add_subplot(1,K_total,k+1)
    ax.scatter(z_sim_final[0,:],z_sim_final[1,:])
    ax.set_title('Latent space of cluster '+str(k+1))
plt.show()
```

 In the above picture, the first (sub-)cluster has been computed anew instead of being copied over. Here are the latent spaces of each stage again one last time!

```python
fig = plt.figure(figsize=(6,6))
ax=fig.add_subplot(1,1,1)
title = ax.set_title("Latent space of all data")
plt.setp(title, color='black') 
for k in range(K_total):
    ax.scatter(z_sim[:,ind==k].T[:,0],z_sim[:,ind==k].T[:,1], c=['red', 'blue', 'green'][k])
plt.show()


fig = plt.figure(figsize=(K_inter*6, 6))
for k in range(K_inter):
    
    W_sim2 = np.mean(returns[k]['W'], axis=0)
    sigma_sim2 = np.mean(returns[k]['sigma'], axis=0)
    mu_sim2 = np.mean(returns[k]['mu'], axis=0)
    z_sim2 = returns[k]['z'][best_inds[k]]
    scnd_ind = ind[z_cats==k]
    
    ax = fig.add_subplot(1,K_inter,k+1)
    for k_i in range(K_total):
        ax.scatter(z_sim2[0,:][scnd_ind==k_i],z_sim2[1,:][scnd_ind==k_i], c=['red', 'green', 'blue'][k_i])
    ax.set_title('Latent space of cluster '+str(k+1))
plt.show()

fig = plt.figure(figsize=(K_total*6, 6))

cluster1_cert = z_sim_norm[0,:][z_sim_norm[0,:]==np.max(z_sim_norm,axis=0)]
cluster2_cert = z_sim_norm3[0,:][z_sim_norm3[0,:]==np.max(z_sim_norm3,axis=0)]
cluster3_cert = z_sim_norm3[1,:][z_sim_norm3[1,:]==np.max(z_sim_norm3,axis=0)]
certs = [cluster1_cert, cluster2_cert, cluster3_cert]

for k in range(K_total):
    W_sim_final = np.mean(returns_final[k]['W'], axis=0)
    sigma_sim_final = np.mean(returns_final[k]['sigma'], axis=0)
    mu_sim_final = np.mean(returns_final[k]['mu'], axis=0)
    z_sim_final = returns_final[k]['z'][best_inds_final[k]]

    ax = fig.add_subplot(1,K_total,k+1)
#     for k_i in range(K_total):
    thrd_ind = ind[z_cats_total==k]
    rgba_colors = np.zeros((len(certs[k]),4))
    rgba_colors[:, 3] = certs[k]**2
    for k_i in range(K_total):
        if k_i==1:
            col=0.5
        else:
            col=1.0
        rgba_colors[thrd_ind==k_i,k_i]=col
    
#         ax.scatter(z_sim[:,ind==k].T[:,0],z_sim[:,ind==k].T[:,1], c=rgba_colors)
#         if k_i == k:
#             print(k_i, k)
    ax.scatter(z_sim_final[0,:],z_sim_final[1,:], c=rgba_colors)
#         else:
#     ax.scatter(z_sim_final[0,:][thrd_ind==k_i],z_sim_final[1,:][thrd_ind==k_i], c=['red', 'blue', 'green'][k_i])
    ax.set_title('Latent space of subcluster '+str(k+1))
plt.show()
```

As suggested in the original article, the density of the ink in each dot represents the posterior probability of the datapoint belonging in the sub-cluster. The colours represent the true cluster in all plots.
