# Mixtures of PPCA's
In this chapter, we introduce mixtures of PPCA models.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pystan
import numpy as np
import seaborn as sns
sns.set()
```

```python
moppcas_code = """

data {
  int<lower=1> K;          // number of mixture components
  int<lower=1> N;          // number of data points
  int<lower=2> D;          // number of observed variables
  int<lower=1> M;          // number of latent dimensions
  matrix[N,D] y;           // observations
}

parameters {
  simplex[K] theta;          // mixing proportions
  vector[D] mu[K];            // locations of mixture components
  vector<lower=0,upper=50>[D] sigma[K];  // scales of mixture components
  matrix[M,N] z[K];  // latent data
  matrix[D,M] W[K];  // factor loadings
  
}

transformed parameters{
    vector[M] mean_z;
    matrix[M,M] cov_z;
    
    for (m in 1:M){
        mean_z[m] = 0.0;
        for (n in 1:M){
            if (m==n){
                cov_z[m,n]=1.0;
            } else{
                cov_z[m,n]=0.0;
            }
        }
    }
}

model {
    vector[K] log_theta = log(theta);  // cache log calculation
    
    for (k in 1:K){
        for (n in 1:N){
            z[k][:,n] ~ multi_normal(mean_z, cov_z);
        }
    }    

    
    for (n in 1:N){
        vector[K] lps = log_theta;
        for (k in 1:K){
            z[k][:,n] ~ multi_normal(mean_z, cov_z);
            lps[k] += normal_lpdf(y[n,:] | W[k]*col(z[k],n)+mu[k], sigma[k]);
            target += log_sum_exp(lps);
            }
        }
}

generated quantities{
    matrix[N,K] clusters;
    
    for (n in 1:N){
        for (k in 1:K){
            clusters[n,k]=0.0;
        }
    }
    
    for (n in 1:N){
        for (k in 1:K){
            for (d in 1:D){
                //expect = row(W[k],d)*col(z[k],n)+mu[k,d];
                //stddev = sigma[k,d];
                clusters[n,k] += normal_lpdf(y[n,d]|mu[k,d], sigma[k,d]);
                
            }
            clusters[n,k] += log(theta[k]);
        }
    }
    
}


"""
moppcas_model = pystan.StanModel(model_code=moppcas_code)
```

```python
N=1000
M=2
D=3
K=3
mu = np.random.uniform(-35,35,size=(K,D))
sigma = np.random.uniform(0,5,size=(K))
theta = np.random.dirichlet(np.ones(K))
z_underlying = np.random.normal(0,1, size=(K,M,N))

opt = np.random.choice(list(range(K)),p=theta, size=N)
z = []
W = []
for k in range(K):
    num_points = sum(opt==k)
    W.append(np.random.normal(0,sigma[k],size=(D,M)))
    z.append(z_underlying[k,:,:num_points])
    
x = []
for k in range(K):
    x.append(np.matmul(W[k],z[k])+mu[k][np.newaxis].T)

print('theta: ', theta)
print()
for k in range(K):
    print('W_'+str(k)+':\n',W[k])
    print('mu_'+str(k)+': ',mu[k,:])
    print('sigma_'+str(k)+': ',sigma[k])
    print()
```

```python
x_total = np.concatenate([x[k].T for k in range(K)])

fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(1,2,1, projection='3d')
for k in range(K):
    ax.scatter(x[k][0,:],x[k][1,:],x[k][2,:], alpha=0.5, label='cluster '+str(k+1))
ax.set_title('Observed data')
ax.set_xlabel('Gene 1')
ax.set_ylabel('Gene 2')
ax.set_zlabel('Gene 3')
plt.legend()
ax = fig.add_subplot(1,2,2, projection='3d')
ax.scatter(x_total[:,0],x_total[:,1],x_total[:,2], alpha=0.5)
ax.set_title('Observed data (unlabeled)')
ax.set_xlabel('Gene 1')
ax.set_ylabel('Gene 2')
ax.set_zlabel('Gene 3')
plt.show()
```

```python
gmm_dat = {'N': N, 'M': M, 'K': K, 'D':D, 'y':x_total}
fit = moppcas_model.sampling(data=gmm_dat, chains=1, iter=100)
fit_ext = fit.extract()
best_ind = np.where(fit_ext['lp__']==max(fit_ext['lp__']))[0][0]
```

```python
theta_sim = np.mean(fit_ext['theta'],axis=0)
mu_sim = np.mean(fit_ext['mu'],axis=0)
sigmas_sim = np.mean(fit_ext['sigma'],axis=0)
z_sim = fit_ext['z'][best_ind]
W_sim = np.mean(fit_ext['W'],axis=0)
clusters = np.mean(fit_ext['clusters'],axis=0)

print('Real theta: ',theta)
print('Sample theta: ',theta)
print('Sample theta: ',[sum(opt==k)/N for k in range(K)])
print('Simulated theta: ',theta_sim)
print()
print('actual mu: ',mu)
print('Simulated mu: ',mu_sim)
print()
print('actual sigmas: ',sigma)
print('Simulated sigmas: ',sigmas_sim)
print()
for k in range(K):
    print('Actual WW^T_'+str(k)+': ')
    print(np.matmul(W[k],W[k].T))
    print('Simulated WW^T_'+str(k)+': ')
    print(np.matmul(W_sim[k],W_sim[k].T))
    print()
```

```python
z_cats = []
for n in range(N):
    for k in range(K):
        if clusters[n,k] == max(clusters[n,:]):
            z_cats.append(k)
            break
```

```python
opt_rec = np.random.choice(list(range(K)),p=theta_sim, size=N)
    
x_rec = []
for k in range(K):
    x_rec.append(np.matmul(W_sim[k],z_sim[k][:,:int(N*theta_sim[k])])+mu_sim[k][np.newaxis].T)
x_total_rec = np.concatenate([x_rec[k].T for k in range(K)])
```

```python
clustermap = np.concatenate([np.ones(sum(opt==k))*k for k in range(K)])
clustermap_rec = np.concatenate([np.ones(int(N*theta_sim[k]))*k for k in range(K)])

fig = plt.figure(figsize=(18,6))
ax = fig.add_subplot(1,3,1, projection='3d')
ax.scatter(x_total[:,0],x_total[:,1],x_total[:,2],c=clustermap, cmap='viridis', alpha=0.5)
ax.set_title('Observed data (Actual clusters)')
ax.set_xlabel('Gene 1')
ax.set_ylabel('Gene 2')
ax.set_zlabel('Gene 3')

ax = fig.add_subplot(1,3,2, projection='3d')
ax.scatter(x_total[:,0],x_total[:,1],x_total[:,2],c=z_cats, cmap='viridis', alpha=0.5)
ax.set_title('Observed data (Guessed Clusters)')
ax.set_xlabel('Gene 1')
ax.set_ylabel('Gene 2')
ax.set_zlabel('Gene 3')

ax = fig.add_subplot(1,3,3, projection='3d')
ax.scatter(x_total_rec[:,0],x_total_rec[:,1],x_total_rec[:,2],c=clustermap_rec, cmap='viridis', alpha=0.5)
ax.set_title('Observed data (Recreated data)')
ax.set_xlabel('Gene 1')
ax.set_ylabel('Gene 2')
ax.set_zlabel('Gene 3')
plt.show()
```

```python
gmm_code = """

data {
  int<lower=1> K;          // number of mixture components
  int<lower=1> N;          // number of data points
  int<lower=1> D;          // number of variables
  matrix[N,D] y;               // observations
}

parameters {
  simplex[K] theta;          // mixing proportions
  matrix[K,D] mu;            // locations of mixture components
  matrix<lower=0,upper=50>[K,D] sigma;  // scales of mixture components
}

model {
    vector[K] log_theta = log(theta);  // cache log calculation
    for (n in 1:N){
        vector[K] lps = log_theta;
        for (k in 1:K)
            lps[k] += normal_lpdf(y[n,:] | mu[k,:], sigma[k,:]);
            target += log_sum_exp(lps);
        }
}

generated quantities{
    matrix[N,K] z;
    
    for (n in 1:N){
        for (k in 1:K){
            z[n,k]=0.0;
        }
    }
    
    
    for (n in 1:N){
        for (k in 1:K){
            for (d in 1:D){
                z[n,k] += normal_lpdf(y[n,d]|mu[k,d], sigma[k,d]);
            }
            z[n,k] += log(theta[k]);
        }
    }
    
}


"""
gmm = pystan.StanModel(model_code=gmm_code)
```

```python
gmm_dat = {'N': N,'K': K, 'D':D, 'y':x_total}
fit = gmm.sampling(data=gmm_dat, chains=1, iter=100)
fit_ext = fit.extract()
best_ind = np.where(fit_ext['lp__']==max(fit_ext['lp__']))[0][0]
```

```python
theta_sim = fit_ext['theta'][best_ind]
mu_sim = fit_ext['mu'][best_ind]
sigmas_sim = fit_ext['sigma'][best_ind]
z_sim = fit_ext['z'][best_ind]
```

```python
z_cats = []
for n in range(N):
    for k in range(K):
        if z_sim[n,k] == max(z_sim[n]):
            z_cats.append(k)
            break
```

```python
z_sim_rec, y_sim = [], []
for i in range(N):
    opt = np.random.choice(list(range(K)),p=theta_sim)
    z_sim_rec.append(opt)
    y_sim.append(np.random.normal(mu_sim[opt],sigmas_sim[opt]))

y_sim = np.array(y_sim)    
    
fig = plt.figure(figsize=(15,10))

ax = fig.add_subplot(2,2,1, projection='3d')
ax.scatter(np.array(x_total)[:,0], np.array(x_total)[:,1], np.array(x_total)[:,2], c=clustermap, cmap='viridis', alpha=0.6)
ax.set_xlim(min(np.min(x_total[:,0]),np.min(x_total[:,0]))-5,max(np.max(x_total[:,0]),np.max(y_sim[:,0]))+5)
ax.set_ylim(min(np.min(x_total[:,1]),np.min(x_total[:,1]))-5,max(np.max(x_total[:,1]),np.max(y_sim[:,1]))+5)
title = ax.set_title("Original sample")
plt.setp(title, color='black') 

ax = fig.add_subplot(2,2,2, projection='3d')
ax.scatter(np.array(x_total)[:,0], np.array(x_total)[:,1], np.array(x_total)[:,2], cmap='viridis')
ax.set_xlim(min(np.min(x_total[:,0]),np.min(x_total[:,0]))-5,max(np.max(x_total[:,0]),np.max(y_sim[:,0]))+5)
ax.set_ylim(min(np.min(x_total[:,1]),np.min(x_total[:,1]))-5,max(np.max(x_total[:,1]),np.max(y_sim[:,1]))+5)
title = ax.set_title("Unlabeled sample")
plt.setp(title, color='black') 

ax = fig.add_subplot(2,2,3, projection='3d')
ax.scatter(np.array(x_total)[:,0], np.array(x_total)[:,1], np.array(x_total)[:,2], c=z_cats, cmap='viridis', alpha=0.6)
ax.set_xlim(min(np.min(x_total[:,0]),np.min(x_total[:,0]))-5,max(np.max(x_total[:,0]),np.max(y_sim[:,0]))+5)
ax.set_ylim(min(np.min(x_total[:,1]),np.min(x_total[:,1]))-5,max(np.max(x_total[:,1]),np.max(y_sim[:,1]))+5)
title = ax.set_title("Guessed clusters")
plt.setp(title, color='black') 

ax = fig.add_subplot(2,2,4, projection='3d')
ax.scatter(np.array(x_total)[:,0], np.array(x_total)[:,1], np.array(x_total)[:,1], c=z_sim_rec, cmap='viridis', alpha=0.6)
ax.set_xlim(min(np.min(x_total[:,0]),np.min(x_total[:,0]))-5,max(np.max(x_total[:,0]),np.max(y_sim[:,0]))+5)
ax.set_ylim(min(np.min(x_total[:,1]),np.min(x_total[:,1]))-5,max(np.max(x_total[:,1]),np.max(y_sim[:,1]))+5)
title = ax.set_title("Recreated data")
plt.setp(title, color='black') 
plt.show()
```
