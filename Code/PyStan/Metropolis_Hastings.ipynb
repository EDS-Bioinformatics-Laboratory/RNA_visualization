```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def prob_func(x):
    return 16-x**2
```

```python
x_points = np.linspace(-4,4,50)
plt.plot(x_points,list(map(prob_func,x_points)))
plt.show()
```

```python
samples = []
x_old = 1
for i in range(100000):
    while True:
        x_new = np.random.normal(x_old,5)
        alpha = prob_func(x_new)/prob_func(x_old)
        if alpha<np.random.uniform():
#             print(x_new, ' rejected, alpha: ', alpha)
            continue
        else:
            x_old = x_new
            samples.append(x_new)
#             print(x_new, ' accepted, alpha: ', alpha)
            break
```

```python
import seaborn as sns
sns.set()
plt.hist(samples, bins=100, density=True)
plt.plot(np.linspace(-4,4,50),list(map(prob_func,np.linspace(-4,4,50)))/sum(list(map(prob_func,np.linspace(-4,4,50)))))
plt.show()
```
